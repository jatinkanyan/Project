```python
total_reservation = len(df)
```

```python
total_cancellation=len(df[df["is_canceled"]==1])
```

```python
import matplotlib.pyplot as plt
```

```python
labels = ['Canceled', 'Not Canceled']
sizes = [total_cancellation, total_reservation - total_cancellation]
colors = ['red', 'green']
explode = (0.1, 0)  # Explode the 'Canceled' slice

plt.figure(figsize=(6, 6))
plt.pie(sizes,  explode=explode,labels=labels, colors=colors, autopct='%1.1f%%',shadow=True,  startangle=140)
plt.title('Cancellation Rate')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
lead_time_data = df["lead_time"]
a=plt.figure(figsize=(8,6))
plt.hist(lead_time_data,bins=30,edgecolor="k",alpha=0.7)
plt.title('Lead Time Distribution')
plt.xlabel('Lead Time (days)')
plt.ylabel('Frequency')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
import seaborn as sns
```

```python
arrival_month=df["arrival_date_month"]
plt.figure(figsize=(10,8))
sns.set(style="darkgrid")
sns.countplot(x=arrival_month,palette="Set2",order=arrival_month.value_counts().index)
plt.title("hotel_arrivals_by_month")
plt.xlabel("month")
plt.ylabel("number of arrivals")
plt.xticks(rotation=45)
plt.show()
```

```python
weekend_nights = df['stays_in_weekend_nights']
week_nights = df['stays_in_week_nights']
plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")
sns.histplot(data=df, x=weekend_nights, bins=20, kde=True, color='skyblue', label='Weekend Nights', alpha=0.7)
sns.histplot(data=df, x=week_nights, bins=20, kde=True, color='salmon', label='Week Nights', alpha=0.7)
plt.title('Distribution of Stays in Weekend Nights and Week Nights')
plt.xlabel('Number of Nights')
plt.ylabel('Frequency')
plt.legend()
sns.despine(left=True, bottom=True)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()







```

```python
adults= df["adults"]
children = df["children"]
babies = df["babies"]
fig, axes = plt.subplots(3, 1, figsize=(10, 12))
sns.set(style="whitegrid")
sns.histplot(data=df,x = adults,bins=10,kde=True,color="skyblue",ax=axes[0])
sns.histplot(data=df,x = children,bins=10,kde=True,color="salmon",ax=axes[1])
sns.histplot(data=df,x = babies,bins=10,kde=True,color="green",ax=axes[2])
axes[0].set_title('Distribution of Adults in Reservations')
axes[1].set_title('Distribution of Children in Reservations')
axes[2].set_title('Distribution of Babies in Reservations')
for ax in axes:
    ax.set_xlabel('Number')
    ax.set_ylabel('Frequency')
sns.despine(left=True, bottom=True)
plt.tight_layout()

plt.show()

```

```python
cancellation_rates = df.groupby('country')['is_canceled'].mean().reset_index()

# Sort the data by cancellation rate in descending order
cancellation_rates = cancellation_rates.sort_values(by='is_canceled', ascending=False).head(10)  # Display the top 10 countries

# Create a bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='is_canceled', y='country', data=cancellation_rates, palette='viridis')
plt.title('Top 10 Countries with the Highest Cancellation Rates')
plt.xlabel('Cancellation Rate')
plt.ylabel('Country')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a pivot table to count reservation statuses for each parking space requirement
pivot_table = df.pivot_table(index='required_car_parking_spaces', columns='reservation_status', aggfunc='size', fill_value=0)
print(pivot_table)

# Create a stacked bar plot
plt.figure(figsize=(10, 6))
pivot_table.div(pivot_table.sum(1), axis=0).plot(kind='bar', stacked=True)
plt.title('Reservation Status vs. Required Car Parking Spaces')
plt.xlabel('Required Car Parking Spaces')
plt.ylabel('Proportion')
plt.xticks(rotation=0)
plt.legend(title='Reservation Status')
plt.show()
```

```python
df['total_stay'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']
average_stay_by_parking = df.groupby('required_car_parking_spaces')['total_stay'].mean()

# Create a bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x=average_stay_by_parking.index, y=average_stay_by_parking.values)
plt.title('Average Length of Stay vs. Required Car Parking Spaces')
plt.xlabel('Required Car Parking Spaces')
plt.ylabel('Average Length of Stay (Nights)')
plt.xticks(rotation=0)
plt.show()
```
